The best Classifier and Regressor from the previous modules are taken and trained. The Classifier predicts if a flight will be delayed or not and if yes, the Regressor predicts the delay in minutes. Performance metrics of the Regressor are computed and shown.

```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

global X_clf,y_clf, X_train_clf, X_test_clf, y_train_clf, y_test_clf, X_reg, y_reg, X_train_reg, y_train_reg
global dataframe

dataframe = pd.read_csv("/content/gdrive/My Drive/file.csv")
X_clf = dataframe.drop(["FlightDate","Origin","Dest","ArrDel15","ArrTime","Unnamed: 0"],axis=1)
y_clf = dataframe["ArrDel15"].tolist()
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf,y_clf, test_size=0.2, random_state=0)
training_data=pd.DataFrame(X_train_clf, columns=X_clf.columns)
training_data["Y"]=y_train_clf
training_data=training_data[training_data["Y"]==1]
X_train_reg=training_data.drop(["Y","ArrDelayMinutes"],axis=1).to_numpy()
y_train_reg=training_data["ArrDelayMinutes"].tolist()
X_train_clf=X_train_clf.drop(["ArrDelayMinutes","Y"],axis=1)
X_test_clf=X_test_clf.drop(["ArrDelayMinutes"],axis=1)
smote = SMOTE()
X_train_clf, y_train_clf = smote.fit_resample(X_train_clf,y_train_clf)
```

```
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import confusion_matrix,classification_report,mean_absolute_error, mean_squared_error
from math import sqrt

#classifier_training
global random_classifier
random_classifier = RandomForestClassifier(random_state=0, n_jobs=-1)
random_classifier = random_classifier.fit(X_train_clf, y_train_clf)
print ("Classifier Trained")


#regressor_training
global random_regressor
random_regressor = RandomForestRegressor(random_state=0, n_jobs=-1)
random_regressor = random_regressor.fit(X_train_reg, y_train_reg)
print ("Regressor Trained")

global y_pred_random_clf
y_pred_random_clf = random_classifier.predict(X_test_clf)
print ("Classifier Predicted")
```

```
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt 

testing_data=pd.DataFrame(X_test_clf)
testing_data["Y_predicted"]=y_pred_random_clf
testing_data=testing_data[testing_data["Y_predicted"]==1]
X_test_reg=testing_data.drop(["Y_predicted"], axis=1)
y_test_reg= X_clf.iloc[np.array(X_test_reg.index),X_clf.columns.get_loc("ArrDelayMinutes")]
y_pred_random_reg = random_regressor.predict(X_test_reg)

score_reg = random_regressor.score(X_test_reg, y_test_reg)
mae_reg= mean_absolute_error(y_test_reg,y_pred_random_reg)
rmse_reg = sqrt(mean_squared_error(y_test_reg, y_pred_random_reg))
print ("Regressor report: ")
print ("Root mean squared error for Random Forest Regresor:",rmse_reg)
print ("Mean Absolute Error for Random Forest Rergressor:",mae_reg)
print ("Score of the Random Forest Regressor: ",score_reg)
```

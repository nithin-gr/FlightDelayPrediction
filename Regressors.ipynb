```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split


global X,y, X_train, X_test, y_train, y_test

dataframe = pd.read_csv("/content/gdrive/My Drive/MLProject - FlightDelay - Apr2020/file.csv")
dataframe = dataframe[dataframe["ArrDel15"]==1]
X = dataframe.drop(["FlightDate","Origin","Dest","ArrDel15","ArrDelayMinutes","ArrTime","Unnamed: 0"],axis=1)
y = dataframe["ArrDelayMinutes"].tolist()
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2, random_state=0)



print (test_set_1)
```

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt 

scaler = StandardScaler()
X_train_a=scaler.fit_transform(X_train)
X_test_a=scaler.transform(X_test)

lin_reg=LinearRegression()
lin_reg = lin_reg.fit(X_train_a,y_train)
y_pred_lin = lin_reg.predict(X_test_a)
score_lin = lin_reg.score(X_test_a, y_test)
mae_lin = mean_absolute_error(y_test,y_pred_lin)
rmse_lin = sqrt(mean_squared_error(y_test, y_pred_lin))
print ("Root mean squared error for Linear Regresor:",rmse_lin)
print ("Mean Absolute Error for Linear Rergressor:",mae_lin)
print ("Score of the Linear Regressor: ",score_lin)
```

```
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt 


dt_reg=DecisionTreeRegressor()
dt_reg = dt_reg.fit(X_train,y_train)
y_pred_dt = dt_reg.predict(X_test)
score_dt = dt_reg.score(X_test, y_test)
mae_dt= mean_absolute_error(y_test,y_pred_dt)
rmse_dt = sqrt(mean_squared_error(y_test, y_pred_dt))
print ("Root mean squared error for Decision Tree Regresor:",rmse_dt)
print ("Mean Absolute Error for Decision Tree Rergressor:",mae_dt)
print ("Score of the Decision Tree Regressor: ",score_dt)
```

# Classification
Different classifiers are trained with the undersampled and oversampled training data and the values are predicted on the testing set. Performance Metrics like 'Score' and the comprehensive Classification Report was show for all the classifiers.

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE

global X,y, X_train, X_test, y_train, y_test

dataframe = pd.read_csv("/content/gdrive/My Drive/file.csv")
X = dataframe.drop(["FlightDate","Origin","Dest","ArrDel15","ArrDelayMinutes","ArrTime","Unnamed: 0"],axis=1)
y = dataframe["ArrDel15"].tolist()
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2, random_state=0)

rus = RandomUnderSampler()
smote = SMOTE()
X_train1, y_train1 = rus.fit_resample(X_train, y_train)
X_train2, y_train2 = smote.fit_resample(X_train,y_train)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report

scaler = StandardScaler()
X_train1a = scaler.fit_transform(X_train1)
X_test1 = scaler.transform(X_test)

log_classifier1 = LogisticRegression(random_state=0, solver='lbfgs', max_iter=2000)
log_classifier1= log_classifier1.fit(X_train1a, y_train1)
y_pred_log1=log_classifier1.predict(X_test1)
confmat_log1 = confusion_matrix(y_test, y_pred_log1)
classrep_log1=classification_report(y_test, y_pred_log1)
print (classrep_log1)
log_score1=log_classifier1.score(X_test1, y_test)
print ("Score (accuracy) of logistic regressor(UnderSampled): ",log_score1)
print ("Confusion Matrix for Logistic Regresor(Undersampled)",confmat_log1)

log_classifier2 = LogisticRegression(random_state=0, solver='lbfgs', max_iter=2000)
log_classifier2= log_classifier2.fit(X_train2, y_train2)
y_pred_log2=log_classifier2.predict(X_test1)
confmat_log2 = confusion_matrix(y_test, y_pred_log2)
classrep_log2=classification_report(y_test, y_pred_log2)
print (classrep_log2)
log_score2=log_classifier2.score(X_test1, y_test)
print ("Score (accuracy) of logistic regressor (Oversampled): ",log_score2)
print ("Confusion Matrix for Logistic Regresor(Oversampled)",confmat_log2)
```

```
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,classification_report

global y_tree_pred

tree_classifier1 = DecisionTreeClassifier(random_state=0)
tree_classifier1= tree_classifier1.fit(X_train1, y_train1)
y_pred_tree1 = tree_classifier1.predict(X_test)
confmat_tree1 = confusion_matrix(y_test, y_pred_tree1)
classrep_tree1 = classification_report(y_test, y_pred_tree1)
print (classrep_tree1)
tree_score1 = tree_classifier1.score(X_test, y_test)
print ("Confusion Matrix for Decision Trees CLassifier(Undersampled)",confmat_tree1)
print ("Score (accuracy) for Decision Trees Classifier(Undersampled) ", tree_score1)

tree_classifier2 = DecisionTreeClassifier(random_state=0)
tree_classifier2= tree_classifier2.fit(X_train2, y_train2)
y_pred_tree2 = tree_classifier2.predict(X_test)
confmat_tree2 = confusion_matrix(y_test, y_pred_tree2)
classrep_tree2 = classification_report(y_test, y_pred_tree2)
print (classrep_tree2)
tree_score2 = tree_classifier2.score(X_test, y_test)
print ("Confusion Matrix for Decision Trees CLassifier(OverSampled)",confmat_tree2)
print ("Score (accuracy) for Decision Trees Classifier(OverSampled) ", tree_score2)
```
